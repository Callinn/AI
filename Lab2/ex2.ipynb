```python
from PIL import Image
```

```python
directory = "data/images"

# Desired image size
image_size = (128, 128)

# Load and resize images
images = []
for filename in os.listdir(directory):
    if filename.endswith((".jpg", ".webp", ".png")):
        img_path = os.path.join(directory, filename)
        img = Image.open(img_path)
        img = img.resize(image_size)
        images.append(img)

# Determine the grid size for displaying images
num_images = len(images)
grid_size = int(num_images**0.5) + 1

# Create a figure with subplots
fig, axes = plt.subplots(grid_size, grid_size, figsize=(10, 10))

# Add images to the subplots
for i, img in enumerate(images):
    row = i // grid_size
    col = i % grid_size
    axes[row, col].imshow(img)
    axes[row, col].axis('off')

# Remove empty subplots
for j in range(i + 1, grid_size * grid_size):
    row = j // grid_size
    col = j % grid_size
    fig.delaxes(axes[row, col])

plt.tight_layout()
plt.show()
```

```python
images = []
for filename in os.listdir(directory):
    if filename.endswith((".jpg", ".webp", ".png")):
        img_path = os.path.join(directory, filename)
        img = Image.open(img_path)
        img = img.resize(image_size)
        img = img.convert("L")  # Convert image to grayscale
        images.append(img)

# Determine the grid size for displaying images
num_images = len(images)
grid_size = int(num_images**0.5) + 1

# Create a figure with subplots
fig, axes = plt.subplots(grid_size, grid_size, figsize=(10, 10))

# Add images to the subplots
for i, img in enumerate(images):
    row = i // grid_size
    col = i % grid_size
    axes[row, col].imshow(img, cmap="gray")
    axes[row, col].axis('off')

# Remove empty subplots
for j in range(i + 1, grid_size * grid_size):
    row = j // grid_size
    col = j % grid_size
    fig.delaxes(axes[row, col])

plt.tight_layout()
plt.show()
```

```python
from PIL import ImageFilter

image_name = "YOLO.jpg"
image_path = os.path.join(directory, image_name)

# Open the image
original_img = Image.open(image_path)

# Apply blur filter to the image
blurred_img = original_img.filter(ImageFilter.GaussianBlur(5))

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Display the original image
axes[0].imshow(original_img)
axes[0].set_title("Before")
axes[0].axis('off')

# Display the blurred image
axes[1].imshow(blurred_img)
axes[1].set_title("After")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python

# Path to the directory containing images
directory = "data/images"

# Image to process
image_name = "YOLO.jpg"
image_path = os.path.join(directory, image_name)

# Open the image using PIL
original_img = Image.open(image_path)

# Apply edge detection filter to the image
edges_img = original_img.filter(ImageFilter.FIND_EDGES)

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Display the original image
axes[0].imshow(original_img)
axes[0].set_title("Before")
axes[0].axis('off')

# Display the image with edges detected
axes[1].imshow(edges_img)
axes[1].set_title("After")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
